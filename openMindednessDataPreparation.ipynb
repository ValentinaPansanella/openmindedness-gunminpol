## Creating dataframe

##### functions to adjust dataframe

```python
datasetName = "openMindednessData"
with open(f"{datasetName}.pickle", "rb") as infile: 
    data = pickle.load(infile)
```

##### dataframe

```python
with open('timestamps.pickle', 'rb') as ifile:
    timestamps = pickle.load(ifile)
```

```python
newtimestamps = dict()
for t in timestamps.keys():
    interval = timestamps[t].split('_')
    data1 = interval[0].split('-')
    firstmonth = data1[1]
    data2 = interval[1].split("-")
    secondmonth = data2[1]
    secondyear = data2[2]
    newinterval = f"{firstmonth}-{secondmonth}/{secondyear}"
    newtimestamps[t] = newinterval
newtimestamps
```

```python
reversetimestamps = {v: k for k, v in newtimestamps.items()}
reversetimestamps
```

```python
df=pd.DataFrame.from_dict({(dataset_name,timestamp,node): data[dataset_name][timestamp][node]
                           for dataset_name in data.keys()
                           for timestamp in data[dataset_name].keys() 
                           for node in data[dataset_name][timestamp].keys()}, 
                           orient='index')

df.reset_index(level=0, inplace=True)
# df.reset_index(level=0, inplace=True)
# df.reset_index(level=0, inplace=True)

df = df.rename(columns={"index":"node", "level_0":"timestamp"})
```
