## Creating dataframe

##### functions to adjust dataframe

```python
datasetName = "openMindednessData"
with open(f"{datasetName}_h.pickle", "rb") as infile: 
    data = pickle.load(infile)
```

##### dataframe

```python
timestamps = {'2017-01-01_2017-07-01':0, '2017-07-01_2018-01-01':1, '2018-01-01_2018-07-01':2, '2018-07-01_2019-01-01':3, '2019-01-01_2019-07-01':4}
newtimestamps = {'01-07/2017':0, '07-12/2017':1, '01-07/2018':2, '07-12/2018':3, '01-07/2019':4}
```

```python
reversetimestamps = {v: k for k, v in newtimestamps.items()}
reversetimestamps
```

```python
df=pd.DataFrame.from_dict({(dataset_name,timestamp,node): data[dataset_name][timestamp][node]
                           for dataset_name in data.keys()
                           for timestamp in data[dataset_name].keys() 
                           for node in data[dataset_name][timestamp].keys()}, 
                           orient='index')

df.reset_index(level=0, inplace=True)
df = df.drop(["level_0"], axis=1)
```

```python
df.reset_index(level=0, inplace=True)
df = df.rename(columns={"level_0":"timestamp"})
```

```python
df.reset_index(level=0, inplace=True)
df = df.rename(columns={"index":"node"})
```

```python
df
```

```python
df['date'] = df['timestamp'].apply(lambda x: reversetimestamps[timestamps[x]])
```

```python
df = df.drop(['timestamp'], axis=1)
df
```
